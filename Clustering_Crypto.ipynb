```python
 # Drop the 'CoinName' column since it's not going to be used on the clustering algorithm
crypto_dataframe = crypto_dataframe.drop("CoinName", axis=1)
print(crypto_dataframe.shape)
crypto_dataframe.head(10)
```

```python
# Create dummy variables for text features
X = pd.get_dummies(data=crypto_dataframe, columns=["Algorithm", "ProofType"])
print(X.shape)
X.head(10)
```

```python
# Standardize data
X = StandardScaler().fit_transform(X)
X[:5]
```

```python
# Use PCA to reduce dimension to 3 principal components
n_comp = 3
pca = PCA(n_components=n_comp)
principal_components = pca.fit_transform(X)
principal_components
```

```python
# Create a DataFrame with the principal components data
col_names = [f"PC {i}" for i in range(1, n_comp + 1)]
pcs_df = pd.DataFrame(principal_components, columns=col_names, index=crypto_df.index)
print(pcs_df.shape)
pcs_df.head()
```

```python
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pcs_df)
    inertia.append(km.inertia_)

# Create the Elbow Curve using altair
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
alt.Chart(df_elbow).mark_line().encode(x="k", y="inertia")
```

```python
# Initialize the K-Means model
model = KMeans(n_clusters=4, random_state=0)

# Fit the model
model.fit(pcs_df)

# Predict clusters
predictions = model.predict(pcs_df)

# Create a new DataFrame including predicted clusters and cryptocurrencies features
clustered_dataframe = pd.concat([crypto_dataframe, pcs_df], axis=1, sort=False)
clustered_dataframe["CoinName"] = coin_name["CoinName"]
clustered_dataframe["Class"] = model.labels_
print(clustered_dataframe.shape)
clustered_dataframe.head(10)
```
